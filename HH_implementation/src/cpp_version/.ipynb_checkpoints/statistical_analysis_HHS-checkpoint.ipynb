```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## General Firing pattern 

```python
Data_28 = pd.read_csv('Data_28.7_HHS1.txt', sep=" ",header=None)
```

```python
time=np.arange(0,105,5).tolist()
frequency=[]
for i in range(0,len(time)):
    frequency.append(Data_28[(Data_28[1]<=time[i]*1000) & (Data_28[1]>time[i-1]*1000)][0].sum())
```
